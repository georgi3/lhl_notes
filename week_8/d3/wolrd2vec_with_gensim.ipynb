```python
import bs4 as bs
import lxml
import re
import nltk
from urllib import request
import gensim
```

```python
scraped_data = request.urlopen('https://en.wikipedia.org/wiki/Artificial_intelligence')
article = scraped_data.read()
```

```python
parsed_article = bs.BeautifulSoup(article, 'lxml')
```

```python
paragraphs = parsed_article.find_all('p')
```

```python
paragraphs
```

```python
article_text = ''
for par in paragraphs:
    article_text += par.text
```

```python
text = article_text.lower()
```

```python
text = re.sub(r'[^a-zA-Z]', ' ', text)
```

```python
text = re.sub(r'\s+', ' ', text)
```

```python
all_sentences = nltk.sent_tokenize(text)
```

```python
all_words = [nltk.word_tokenize(sent) for sent in all_sentences]
```

```python
len(all_words)
```

```python
for i in range(len(all_words)):
    all_words[i] = [word for word in all_words[i] if word not in nltk.corpus.stopwords.words('english')]
```

```python
len(all_words[0])
```

```python
word2vec = gensim.models.Word2Vec(all_words, min_count=2)
```

```python
vocabulary = word2vec.wv.index_to_key
```

```python
vocabulary
```
