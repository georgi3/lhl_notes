```python
x = np.linspace(0, 40, 200)
y = np.sin(x)
y1 = (np.sin(x))*(np.cos(x))
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, c='r')
ax.plot(x, y1, c='g')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 6))
ax[0].plot(x, y)
ax[0].set_title('sin(x)')

ax[1].plot(x, y1, c='green')
ax[1].set_title('(np.sin(x))*(np.cos(x))')
plt.tight_layout()
```

## Legend

```python
iris = datasets.load_iris()
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df
```

```python
iris_df['target_name'] =iris_df.target.map(
    {0: 'setosa',
     1: 'versicolor',
     2: 'virginica'}
)
iris_df.head()
```

```python
setosa = iris_df[iris_df['target_name'] == 'setosa']
versicolor = iris_df[iris_df['target_name'] == 'versicolor']
virginica = iris_df[iris_df['target_name'] == 'virginica']
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.scatter(setosa['sepal length (cm)'], setosa['sepal width (cm)'], marker='o', color='red',
           label='setosa')
ax.scatter(versicolor['sepal length (cm)'], virginica['sepal width (cm)'], marker='o', color='green',
           label='versicolor')
ax.scatter(virginica['sepal length (cm)'], virginica['sepal width (cm)'], marker='o', color='blue',
           label='virginica')

# Annotations
ax.annotate('setosa', xy=(5.0, 3.5),
            xytext=(4.25,4.0), arrowprops={'color': 'red'})
ax.annotate('virginica', xy=(7.2,3.6),
            xytext = (6.5,4.0), arrowprops={'color': 'blue'})
ax.annotate('versicolor', xy=(5.33, 2.61),
            xytext = (4.5,2.75), arrowprops={'color': 'green'})

ax.legend(loc='upper right')
ax.set_title('Iris flower')
ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_xlim(4,8)
```
