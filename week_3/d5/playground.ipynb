```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (12,8)
```

```python
l = [[3, 4], [3, 5], [0, 5], [0, 0], [1, 0], [1, 1.8], [2.5, 1.8], [2.5, 2.8], [1, 2.8], [1, 4], [3, 4]]
nl = np.array(l)
```

```python
plt.plot(nl[:, 0], nl[:, 1])
```

```python
transformed = [(0.4 * p[0] + p[1], p[0] + 0 * p[1]) for p in l]
tr = np.array(transformed)
```

```python
plt.plot(nl[:, 0], nl[:, 1])
plt.plot(tr[:, 0], tr[:, 1]);
```

```python
transformed = [(np.cos(45) * p[0] + 1 * p[1],
                np.cos(90) * p[0] + 0 * p[1])
               for p in l]
tr = np.array(transformed)
```

```python
plt.plot(nl[:, 0], nl[:, 1])
plt.plot(tr[:, 0], tr[:, 1]);
```
